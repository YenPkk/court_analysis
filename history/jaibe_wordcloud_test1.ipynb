```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import jieba
import jieba.analyse
from collections import Counter # 次數統計
```

```python
import pandas as pd
```

```python
case500 = pd.read_csv('./cases_500_v4.csv')
```

```python
context_test = case500['context']
```

```python
word_count = {}
# for i in str_list:
#     if i in word_count:
#         word_count[i] +=1
#     else:
#         word_count[i] = 1

for d in context_test:
    str_list = jieba.cut(d)
    for i in str_list:
        if i in word_count:
            word_count[i] +=1
        else:
            word_count[i] = 1
```

```python
with open ('jieba.txt', 'w', encoding='utf-8' ) as f:
    f.write(str(word_count))
```

```python
for i in context_test[:1]:
    test_02 = i
```

```python
len(test_02)
```

```python
seg_list = jieba.lcut(test_02, cut_all=True)
```

```python
dictionary = Counter(seg_list)
```

```python
STOP_WORDS = [' ', '，', '（', '）', '...', '。', '「', '」', '[', ']','\n']
[dictionary.pop(x, None) for x in STOP_WORDS] # 從字典裡刪除停用詞
print(dictionary) # 把計算完的每個分詞出現次數顯示出來看看
```

```python
font_path = 'NotoSansTC-VariableFont_wght.ttf'
wc = WordCloud(background_color='white',
                         margin=2, # 文字間距
                         font_path=font_path, # 設定字體
                         max_words=200, # 取多少文字在裡面
                         width=1280, height=720) # 解析度
```

```python
wc.generate_from_frequencies(dictionary) # 吃入次數字典資料
```
